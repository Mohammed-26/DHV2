```python
import numpy as np
import matplotlib.pyplot as plt

# Create grid coordinates
x = np.linspace(-2, 2, 10)
y = np.linspace(-2, 2, 10)
X, Y = np.meshgrid(x, y)

# Define vector components
U = np.sin(X) * np.cos(Y)
V = np.cos(X) * np.sin(Y)

# Create quiver plot
plt.figure()
plt.title('Quiver Plot with Multiple Arrows Forming a Pattern')
plt.quiver(X, Y, U, V)
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create grid coordinates
x = np.linspace(-2, 2, 10)
y = np.linspace(-2, 2, 10)
X, Y = np.meshgrid(x, y)

# Define vector components
U = np.sin(X) * np.cos(Y)
V = np.cos(X) * np.sin(Y)

# Calculate the magnitude of each arrow
speed = np.sqrt(U**2 + V**2)

# Create quiver plot with colored arrows
plt.figure()
plt.title('Quiver Plot with Colorful Arrows')
plt.quiver(X, Y, U, V, speed, cmap='viridis', scale=20)
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.colorbar(label='Magnitude')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define grid coordinates
x = np.linspace(-2, 2, 20)
y = np.linspace(-2, 2, 20)
X, Y = np.meshgrid(x, y)

# Define vector components for curved arrows forming a pattern
U = Y * np.sin(X)
V = -X * np.cos(Y)

# Create quiver plot with curved arrows
plt.figure()
plt.title('Quiver Plot with Curved Arrows Forming a Dragon Pattern')
plt.quiver(X, Y, U, V, scale=30, headaxislength=4)
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define grid coordinates
x = np.linspace(-2, 2, 20)
y = np.linspace(-2, 2, 20)
X, Y = np.meshgrid(x, y)

# Define parameters for the radial map
r = np.sqrt(X**2 + Y**2)
theta = np.arctan2(Y, X)

# Define vector components for the radial map pattern
U = r * np.cos(theta)**3
V = r * np.sin(theta)**3

# Create quiver plot with curved arrows forming a radial map
plt.figure()
plt.title('Quiver Plot with Complex Radial Map Pattern')
plt.quiver(X, Y, U, V, scale=30, headaxislength=4)
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.show()
```
